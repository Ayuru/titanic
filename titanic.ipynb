```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

input_path = Path(r"C:\Users\ayuru\Desktop\Py\titanic\titanic.csv")
titanic = pd.read_csv(input_path)
```

```python
mean = titanic["Fare"].mean()
print(f"Średnia cena biletu wynosi około {mean:.2f}")
```

```python
quantiles = titanic["Fare"].quantile([0.25, 0.5, 0.75])
print(f"Kwantyle wynoszą:\n{quantiles}")
```

```python
range_fare = max(titanic["Fare"]) - min(titanic["Fare"])
print(f"zasięg wynosi: {range_fare:.2f}")
```

```python
variance = np.var(titanic["Fare"], ddof=1)
print(f"Wariancja wynosi {variance:.3f}, a odchylenie standardowe {variance**(1/2):.3f}")
```

```python
asymetry = pd.Series(titanic["Fare"]).skew()
print(f"Asymetria rozkałdu wynosi: {asymetry}")
```

```python
plt.figure(figsize = (15, 8))
titanic["Fare"].plot(kind='hist',bins=20, edgecolor='black')
plt.title("Histogram")
plt.xlabel('Cena ', fontsize=16, fontweight="bold")
plt.ylabel('Liczba biletów', fontsize=16, fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.boxplot(titanic["Fare"].dropna())
plt.title("Wykres pudełkowy")
plt.ylabel("Fare")
plt.show()
```
